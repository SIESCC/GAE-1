#Tokenization

```python
# Sentence tokenization for Japanese requires a specialized library.
!pip install janome
from janome.tokenizer import Tokenizer

def janome_sentence_tokenize(text):
    sentences = []
    current_sentence = []
    t = Tokenizer()
    for token in t.tokenize(text):
        current_sentence.append(token.surface)
        # Common sentence-ending punctuation in Japanese
        if token.surface in ['。', '？', '！']:
            sentences.append(''.join(current_sentence))
            current_sentence = []
    if current_sentence:
        sentences.append(''.join(current_sentence))
    return sentences

sentences = janome_sentence_tokenize(text)
print("Sentences:", sentences[:5])
```

```python
# Word tokenization for Japanese requires a specialized library (morphological analyzer).
from janome.tokenizer import Tokenizer

t = Tokenizer()
words = [token.surface for token in t.tokenize(text)]
print("Words:", words[:20])
```

```python
# Sample counts
print("Sentence count:", len(sentences))
print("Word count:", len(words))
```

## Frequency Distribution

```python
fdist = FreqDist(words)
print(fdist)                      # FreqDist object
print(fdist.most_common(10))      # Top 10 most frequent
```

```python
# Graph: Frequency Plot
# Plot top 20 most frequent tokens
# This is a font issue, not a code logic issue.

plt.figure(figsize=(12,6))
fdist.plot(20, title="Top 20 Most Frequent Tokens")
plt.show()
```

## Conditional Frequency Distribution

```python
cfd = ConditionalFreqDist(
    (len(word), word.lower())
    for word in words
    if word.isalpha()
)
```

```python
# Show counts for 3 selected lengths
print("Words of length 3:", cfd[3].most_common(10))
print("Words of length 4:", cfd[4].most_common(10))
```

```python
# Plot CFD Graph
plt.figure(figsize=(12,6))
cfd.plot(conditions=[2,3,4,5,6],
         title="Conditional Frequency Dist: Word lengths vs tokens")
plt.show()
```

## Vocabulary Analysis

```python
# Vocabulary = unique set of tokens
vocab = set(words)
print("Vocabulary size:", len(vocab))

# Sorted vocabulary
vocab_sorted = sorted(vocab)
print(vocab_sorted[:20])
```

```python
# Longest words
long_words = [w for w in vocab if len(w) > 3]
print("Words longer than 10 characters:", long_words)
```

## Concordance & Collocations

```python
from nltk.text import Text

nltk_text = Text(words)

# Concordance lookup
nltk_text.concordance("雨")
```

```python
# Multiple hits
nltk_text.concordance("私")
```

```python
# Collocations (Common pairs)
nltk_text.collocations()
```

## Regular Expressions

```python
import re

# All Japanese character tokens
jp_words = [w for w in vocab if re.fullmatch(r"[ぁ-んァ-ン一-龯]+", w)]
print("Japanese tokens only:", jp_words[:20])
```

```python
import re

# Words ending with 「た」
ta_words = [w for w in vocab if re.search(r"た$", w)]
print("Words ending with 'た':", ta_words[:20])
```

```python
import re

# Words starting with Katakana
katakana_words = [w for w in vocab if re.search(r"^[ァ-ン]", w)]
print("Katakana tokens:", katakana_words[:20])
```

```python
import re

# Find tokens matching patterns
pattern_results = [w for w in vocab if re.search(r"雨.+", w)]
print("Pattern matches for '雨…':", pattern_results)

```
